Electrical formulas used:

• Series resistance: Rₛ = R₁ + R₂ + …

• Parallel resistance: 1/Rₚ = 1/R₁ + 1/R₂ + …

• Ohm’s Law: V = I × R

Syntax

• print() for output
          
• input() for user input
          
• float() for numeric conversion

• Arithmetic operators: + − * /


Program to calculate equivalent resistance in series

```python
r1=int(input("r1="))
r2=int(input("r2="))
rs=r1+r2
print("rs=",rs)
```

Program to calculate equivalent resistance in parallel

```python
r1=float(input("r1="))
r2=float(input("r2="))
rp=1/(1/r1)+(1/r2)
print(rp)
```

Program to calculate voltage using Ohm’s Law

```python
i=float(input("i="))
r=float(input("r="))
v=i*r
print(v,"ohms")
```
